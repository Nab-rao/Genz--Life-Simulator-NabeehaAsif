```python
# ğŸ€ Assignment 1 â€“ Personal AI Chatbot (AIC-202)
# Project Title: Basketball Life Simulator

import random

# 1. Introduction
def chatbot_intro():
    print("Yo! ğŸ‘‹ Welcome to the Basketball Life Simulator!")
    name = input("Whatâ€™s your name, baller? ğŸ€: ")
    age = input("How old are you?: ")
    print(f"Nice to meet you, {name}! At {age}, you're ready to dominate the court ğŸ˜")
    return name

# 2. Basic Chatbot Responses
responses = {
    "hi": "Hey champ! ğŸ€",
    "hello": "Whatâ€™s good baller? ğŸ˜",
    "how are you": "Chillinâ€™ and ready to hoop! What about you?",
    "bye": "Catch you later! Keep grinding ğŸ’ªğŸ€",
    "thanks": "Anytime! Teamwork makes the dream work ğŸ”¥",
    "welcome": "Glad to help! ğŸ™Œ"
}

fallback_messages = [
    "Hmmâ€¦ never heard that play before ğŸ˜…",
    "Say that again, baller?",
    "Interesting moveâ€¦ but try again ğŸ‘€",
    "Try another move, baller ğŸ€",
    "Not sure what that means, but Iâ€™m still here ğŸ˜"
]

def get_response(message):
    msg = message.lower()

    # Priority handling for YES
    if msg in ["yes", "y", "yeah", "sure", "ok"]:
        return "YES_CONFIRM"

    if msg in responses:
        return responses[msg]

    elif "play" in msg:
        return "Wanna start your Basketball Life Simulator? (yes/no)"

    else:
        return random.choice(fallback_messages)

# 3. Player Stats
player = {"energy": 100, "skill": 0, "mood": "ğŸ”¥"}

# â­ NEW SMALL FEATURE: Motivation quotes
quotes = [
    "ğŸ€ 'Hard work beats talent when talent fails to work hard.'",
    "ğŸ”¥ 'You miss 100% of the shots you donâ€™t take.'",
    "ğŸ’ª 'Champions are made when no one is watching.'",
    "âš¡ 'Fall down seven times, get up eight.'",
    "ğŸ¯ 'Push yourself, because no one else will do it for you.'"
]

def give_quote():
    print("\nMotivation Boost:")
    print(random.choice(quotes))

# 4. Basketball Mini Game Feature
def basketball_minigame():
    print("\nğŸ€ Time to ball out! Choose your drill:")
    print("1. Shooting Practice ğŸ¯")
    print("2. Dribbling Drill â±")
    print("3. 1v1 Challenge ğŸ’¥")

    choice = input("Pick an option (1â€“3): ")

    if choice == "1":
        print("You practiced your jump shotâ€¦ swish! ğŸ€âœ¨")
        player["skill"] += 10
        player["energy"] -= 10
        player["mood"] = "Focused ğŸ˜Œ"

    elif choice == "2":
        print("Quick handles! ğŸ”¥ Your dribbling improved!")
        player["skill"] += 15
        player["energy"] -= 15
        player["mood"] = "Sharp âš¡"

    elif choice == "3":
        print("1v1 timeâ€¦ let's see what happens ğŸ‘€")
        outcome = random.choice([
            "You cooked your opponent! ğŸ˜ğŸ”¥",
            "Close game, but you lost ğŸ˜­",
            "You hit a clutch shot at the buzzer! ğŸ€ğŸ’¥"
        ])

        print(outcome)

        if "cooked" in outcome:
            player["skill"] += 20
            player["mood"] = "Confident ğŸ’ª"
        elif "lost" in outcome:
            player["skill"] += 5
            player["energy"] -= 10
            player["mood"] = "Motivated ğŸ˜¤"
        else:
            player["skill"] += 30
            player["mood"] = "Legendary ğŸ‘‘"

    else:
        print("That's not a real basketball drill, baller ğŸ˜…")

    player["energy"] = max(0, min(100, player["energy"]))

    print(f"âš¡ Energy: {player['energy']} | â­ Skill: {player['skill']} | ğŸ˜„ Mood: {player['mood']}")

# 5. Main Activities
def perform_action(action):
    if action == "rest":
        print("You took a break and relaxed ğŸ˜´")
        player["energy"] = 100
        player["mood"] = "Recovered ğŸ’†â€â™‚"

    elif action == "train":
        print("Time to level up your game! ğŸ†")
        basketball_minigame()

    elif action == "stretch":
        print("You stretched and avoided an injury ğŸ™")
        player["energy"] += 10
        player["mood"] = "Loose ğŸ§˜"

    # â­ NEW small feature: motivation quote command
    elif action == "quote":
        give_quote()
        return

    else:
        print("Thatâ€™s not something we can do on the court ğŸ˜…")
        print("Try: rest / train / stretch / quote / quit")
        return

    player["energy"] = max(0, min(100, player["energy"]))
    print(f"âš¡ Energy: {player['energy']} | â­ Skill: {player['skill']} | ğŸ˜„ Mood: {player['mood']}")

# 6. Game Loop
def start_game():
    print("\nğŸ€ Game Started! Let's build your basketball journey ğŸ’ª")
    while True:
        act = input("Choose: rest/train/stretch/quote/quit: ").lower()
        if act == "quit":
            print("\nâœ¨ Game Over âœ¨")
            print(f"Final Stats â†’ âš¡ {player['energy']} | â­ {player['skill']} | ğŸ˜„ {player['mood']}")
            print("Great hustle today! ğŸ€ğŸ”¥")
            break
        perform_action(act)

# 7. Chatbot Loop
def start_chatbot():
    chatbot_intro()
    while True:
        user_msg = input("\nYou: ")
        bot_reply = get_response(user_msg)
        print("Bot:", bot_reply)

        if bot_reply == "YES_CONFIRM":
            start_game()

        elif "wanna start" in bot_reply.lower():
            answer = input("You: ").lower()
            if answer in ["yes", "y", "yeah", "ok", "sure"]:
                start_game()
            else:
                print("No worries baller, warm up anytime ğŸ’¬")

        elif user_msg.lower() == "bye":
            print("Peace out! Keep practicing ğŸ€âœ¨")
            break

# 8. Run Chatbot
start_chatbot()
```
